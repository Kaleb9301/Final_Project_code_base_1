```python
from graphviz import Digraph

dot = Digraph()

# Data Preprocessing
dot.node('A', 'Data Preprocessing')
dot.node('A1', 'Read Dataset')
dot.node('A2', 'Select Relevant Columns')
dot.node('A3', 'Fix Specific Data Points')
dot.node('A4', 'Check for Null Values')
dot.node('A5', 'Drop Rows with Nulls')

# Define the edges
dot.edge('A1', 'A')
dot.edge('A2', 'A')
dot.edge('A3', 'A')
dot.edge('A4', 'A')
dot.edge('A5', 'A')

# Feature Scaling
dot.node('B', 'Feature Scaling')
dot.node('B1', 'Split Data into Features and Labels')
dot.node('B2', 'Scale Features using MinMaxScaler')

# Define the edges
dot.edge('B1', 'B')
dot.edge('B2', 'B')

# Data Preparation for LSTM
dot.node('C', 'Training Data Generator')
dot.node('D', 'Testing Data Generator')

# Model Development
dot.node('E', 'Model Development')
dot.node('E1', 'Define LSTM Model')
dot.node('E2', 'Compile Model')
dot.node('E3', 'Train Model with Training Data')

# Define the edges
dot.edge('E1', 'E')
dot.edge('E2', 'E')
dot.edge('E3', 'E')

# Prediction and Evaluation
dot.node('F', 'Prediction and Evaluation')
dot.node('F1', 'Predict on Test Data')
dot.node('F2', 'Inverse Transform Predictions')
dot.node('F3', 'Visualize Predictions')

# Define the edges
dot.edge('F1', 'F')
dot.edge('F2', 'F')
dot.edge('F3', 'F')

# Connect the main blocks
dot.edge('A', 'B')
dot.edge('B', 'C')
dot.edge('C', 'D')
dot.edge('D', 'E')
dot.edge('E', 'F')

# Render the diagram
dot.render('model_development_process', format='png')
```
